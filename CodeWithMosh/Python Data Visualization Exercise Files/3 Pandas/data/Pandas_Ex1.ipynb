# Creating heatmaps

```python
#Sequential(low to high), Qualittaive (discrete chunks of data that have no ordering) heatmaps are the two types
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
confusion=np.array([[37,0,0,0],
                    [0,39,1,0],
                    [0,3,41,2],
                    [0,0,0,44]])
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(confusion,annot=True,cmap='Blues');
plt.ylabel('Actual Label');
plt.xlabel=('Predicted Label')
```

```python
#For qualititaive
plt.figure(figsize=(6,6))
sns.heatmap(confusion,annot=True,cmap='Pastel1');#Qualitative
plt.ylabel('Actual Label');
plt.xlabel=('Predicted Label')
```

# Creating subplots, useful for comparing two subsets of data

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename='digitsDataset.csv'
df=pd.read_csv(filename)
df.head()
```

```python
pixel_colnames=df.columns[:-1]#all except last column which is our label of the image
pixel_colnames
```

```python
image_values=df.loc[0,pixel_colnames].values
```

```python
image_values.shape
```

```python
image_values.reshape(8,8)
```

```python
plt.imshow(image_values.reshape(8,8),cmap='gray')#Little blurry bcoz it is an 8** image
```

```python
#we will create  5 subplots using .subplot command
plt.figure(figsize=(10,2))
plt.subplot(1,5,1)#1 row, 5 columns, this is 1st of the subplots
image_values=df.loc[0,pixel_colnames].values
image_label=df.loc[0,'label']
plt.imshow(image_values.reshape(8,8),cmap='gray')
plt.title('Label:'+str(image_label))

plt.subplot(1,5,2)#1 row, 5 columns, this is 1st of the subplots
image_values=df.loc[1,pixel_colnames].values
image_label=df.loc[1,'label']
plt.imshow(image_values.reshape(8,8),cmap='gray')
plt.title('Label:'+str(image_label))

plt.subplot(1,5,3)#1 row, 5 columns, this is 1st of the subplots
image_values=df.loc[2,pixel_colnames].values
image_label=df.loc[2,'label']
plt.imshow(image_values.reshape(8,8),cmap='gray')
plt.title('Label:'+str(image_label))

plt.subplot(1,5,4)#1 row, 5 columns, this is 1st of the subplots
image_values=df.loc[3,pixel_colnames].values
image_label=df.loc[3,'label']
plt.imshow(image_values.reshape(8,8),cmap='gray')
plt.title('Label:'+str(image_label))

plt.subplot(1,5,5)#1 row, 5 columns, this is 1st of the subplots
image_values=df.loc[4,pixel_colnames].values
image_label=df.loc[4,'label']
plt.imshow(image_values.reshape(8,8),cmap='gray')
plt.title('Label:'+str(image_label))
```

```python
#Lot of duplication, we can use for loop to eliminate
```

```python
plt.figure(figsize=(10,2))
for index in range(0,5):
    plt.subplot(1,5,1+index)
    image_values=df.loc[index,pixel_colnames].values
    image_label=df.loc[index,'label']
    plt.imshow(image_values.reshape(8,8),cmap='gray')
    plt.title('Label:'+str(image_label))
```
